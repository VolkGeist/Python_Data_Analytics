Staffelter Hof Winery is Germany's oldest business, established in 862 under the Carolingian dynasty. It has continued to serve customers through dramatic changes in Europe, such as the Holy Roman Empire, the Ottoman Empire, and both world wars. What characteristics enable a business to stand the test of time?

To help answer this question, BusinessFinancing.co.uk researched the oldest company still in business in **almost** every country and compiled the results into several CSV files. This dataset has been cleaned.

Having useful information in different files is a common problem. While it's better to keep different types of data separate for data storage, you'll want all the data in one place for analysis. You'll use joining and data manipulation to work with this data and better understand the world's oldest businesses.

## 1. What is the oldest business on each continent?

````python
merged_businesses = pd.merge(businesses, countries, on='country_code')
````

````python
#oldest_business = merged_businesses.loc[merged_businesses.groupby(by = 'continent')['year_founded'].idxmin()]
oldest_business = merged_businesses.loc[merged_businesses.groupby(by = 'continent')['year_founded'].idxmin(),['business','country','continent','year_founded']]
oldest_business
````

## 2. How many countries per continent lack data on the oldest businesses?

````python
business_country = pd.merge(businesses, countries, on='country_code', how='right')
business_country
````

````python
countries_without_business = business_country[business_country.business.isna()]
````

````python
countries_without_business = countries_without_business.groupby(by='continent').agg(countries_without_businesses = ('country','count')).reset_index()
countries_without_business
````

## 3. Does including the `new_businesses` data change this?

````python
businesses_newest = pd.concat([businesses,new_businesses],axis = 0).reset_index(drop = True)
````

````python
business_country_new = pd.merge(businesses_newest, countries, on='country_code', how='right')
countries_without_business_new = business_country_new[business_country_new.business.isna()]
countries_without_business_new = countries_without_business_new.groupby(by='continent').agg(countries_without_businesses_new = ('country','count')).reset_index()
````

````python
countries_without_business.merge(countries_without_business_new, on = 'continent')
````

Yes, it does

## 4. Which business categories are best suited to last over the course of centuries?

````python
business_cat = pd.merge(businesses_newest, categories, on='category_code', how='inner')
````

````python
business_cat_cont = pd.merge(business_cat, countries, on='country_code', how='inner')
````

````python
business_cat_cont.groupby(by=['continent','category']).min('year_founded').reset_index()
````
